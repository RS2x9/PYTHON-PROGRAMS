corr function : also known as corelation function.

it outputs how variables are related to each other

performs only on the numerical data and ignores the numerical datas 

```python
import pandas as pd 
import seaborn as sns
match = pd.read_csv(r"D:\delivery-matches\matches.csv")
match.head(3)
```

don't run the immediate below one 

```python
print(match.corr())    # this causes error
```

reason: if  DataFrame has columns with mixed types (numbers and strings in the same column), pandas may try to convert them to float and fail.

You can select only numeric columns before calling corr():

```python
x= match.select_dtypes( include='number').corr()
x
# it cretes a pivot table and shows  relation between the same columns 
# the realted b/w the same column is always 1
# negative correraltion means datas are inversly proportional 
# positive correlation means directly proportional 
```

```python
sns.heatmap(x)
# light color : strong correlation when variables are directly proportional 
# dark color : strong correlation when variables are inversely proportional 
```
