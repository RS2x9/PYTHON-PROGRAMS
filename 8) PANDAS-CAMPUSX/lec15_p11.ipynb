pivot_table() : it folds the data into other dimensions and gives a clear view 

--> syntax:

pivot_table(index , columns , values , aggfunc)

```python
import pandas as pd 
data= {
    'NAME': ['NITISH', 'ANU', 'MUKKKU', 'SURI', 'RAJIV', 'Vikram', 'Nidhi', 'Aditya', 'Nidhi', 'Gaurav', 'Vikram', 'Nidhi', 'Gaurav', 'Kavita', 'Aditya', 'Anjali', 'Deepak', 'Amit', 'Kavita', 'Mohit'],
    'GENDER': ['M', 'F', 'M', 'M', 'M', 'M', 'F', 'F', 'M', 'M', 'F', 'F', 'M', 'M', 'F', 'F', 'F', 'F', 'F', 'M'],
    'CITY': ['KOLKATA', 'GURGAON', 'KOLKATA', 'KOLKATA', 'PATNA', 'MUMBAI', 'MUMBAI', 'DELHI', 'CHENNAI', 'CHENNAI', 'MUMBAI', 'KOLKATA', 'KOLKATA', 'CHENNAI', 'BANGALORE', 'GURGAON', 'GURGAON', 'MUMBAI', 'PATNA', 'BANGALORE'],
    'FREQUENCY': ['WEEKLY', 'DAILY', 'ONCE', 'MONTHLY', 'NEVER', 'DAILY', 'ONCE', 'MONTHLY', 'DAILY', 'NEVER', 'NEVER', 'MONTHLY', 'NEVER', 'NEVER', 'ONCE', 'DAILY', 'DAILY', 'WEEKLY', 'DAILY', 'MONTHLY'],
    'ITEM': ['BURGER', 'SANDWICH', 'VADA', 'PIZZA', 'PANEER', 'PANEER', 'BIRYANI', 'DOSA', 'CHOWMEIN', 'CHOWMEIN', 'SANDWICH', 'PASTA', 'BIRYANI', 'SANDWICH', 'CHOWMEIN', 'SANDWICH', 'CHOWMEIN', 'CHOWMEIN', 'DOSA', 'CHOWMEIN'],
    'SPENDS': [11, 14, 25, 56, 34, 10, 67, 65, 18, 22, 89, 44, 24, 57, 92, 76, 20, 57, 65, 78]
}
food = pd.DataFrame(data)
food
```

eg: in all the cities , give the mean value of price of items 

```python
x= food.pivot_table(index='CITY' , columns='ITEM' , values ='SPENDS' , aggfunc= 'mean')
x
```

eg: in all the cities , give the mean value of price of items  for both males and females 

```python
x= food.pivot_table(index=['CITY' , 'GENDER'] , columns='ITEM' , values='SPENDS', aggfunc='mean')
x
```

eg: in all the cities , give the mean value of price of items  for both males and females 
on the basis of their eating frequency

```python
x= food.pivot_table(index=['CITY' , 'GENDER'] ,columns= ['ITEM','FREQUENCY'], values='SPENDS' , aggfunc='mean')
x
```
