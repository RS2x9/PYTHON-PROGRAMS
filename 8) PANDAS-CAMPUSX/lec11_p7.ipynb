Groupby() : is a pandas method used to split your DataFrame into groups based on the values in one or more columns.

example :

import pandas as pd

data = {
    'device_id': ['A', 'A', 'B', 'B', 'C'],
    'temperature': [22, 23, 21, 20, 19]
}

df = pd.DataFrame(data)     # creates pandas dataframe of the dictionary data and converts it into  tabular structure 

Now group by device_id and get the average temperature:

df.groupby('device_id')['temperature'].mean()       

result : 

device_id

A           22.5

B           20.5

C           19.0

import pandas as pd

data = {
    'device_id': ['A', 'A', 'B', 'B', 'C'],
    'temperature': [22, 23, 21, 20, 19]
}

df = pd.DataFrame(data)

grouped = df.groupby('device_id')

print(grouped.groups)

Output:

{
 'A': [0, 1],

 'B': [2, 3],
 
 'C': [4]
}


- The groups attribute returns a dictionary.

- Each key is a unique value from the device_id column.

- Each value is a list of row indices in the DataFrame where that key appears.


```python
company.groups
```

- List of indices: df.iloc[[0, 2]] → rows 0 and 2
